```python
import matplotlib.pyplot as plt
dfai["patrimonio"].value_counts().plot.barh(title="Total imóveis por patrimônio", color="orangered")
plt.xlabel("Patrimônio")
plt.ylabel("Total Imóveis");
plt.legend();
```
